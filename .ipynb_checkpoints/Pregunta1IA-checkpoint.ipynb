## Funciones y graficación


```python
point  = np.array([3., 2., 1.])
normal = np.array([-1., -1., -1.])

# La ecuación del plano es a*x+b*y+c*z=d
# Los coeficientes del plano son [a,b,c] y 
# forman lo que se conoce como la normal
d = normal.dot(point)

print("Valor de d:",d)

# En x, y se construye la cuadrícula conmeshgrid 
# inicialmente de dos dimensiones, para calcular z luego
x, y = np.meshgrid(range(10), range(10))

##############################################################
# si a*x + b*y + c*z = d
# c*z = -a*x - b*y + d
# z = ( -a*x - b*y + d ) / c
z = (-normal[0]*x - normal[1]*y + d)  / normal[2]

# Manejo del gráfico
plt.clf()
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(x, y, z,cmap=cm.coolwarm )
ax.set_xlabel("Nuestro X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
#Paraboloide hiperbólica, también se lo conoce bajo los nombres de silla de 
#montar o paso de montaña por su conformación geométrica

def hiperbolica(x, y):
  a = 3
  b = 2
  c = 1
  return c * ( (y**2 / b**2) - (x**2 / a**2) )

x = y = np.arange(-3.0, 3.0, 0.05)
X, Y = np.meshgrid(x, y)

# np.ravel return a contiguous flattened array.
zs = np.array([hiperbolica(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

# Manejo del gráfico
plt.clf()
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z,cmap=cm.coolwarm )
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()

print(zs.shape)
```

```python
def eliptica(x, y):
  a = 3
  b = 2
  c = 1
  return c * ( (x**2 / a**2) + (y**2 / b**2) )

x = y = np.arange(-3.0, 3.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array([eliptica(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

# Manejo del gráfico
plt.clf()
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z,cmap=cm.coolwarm )
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

## El vector gradiente

A continuación se define la función **derivada parcial** de una
función de dos variables $z=f\left(x,y\right)$ con dominio en $\mathbb{R}^{2}$
y codominio en $\mathbb{R}$ respecto a $x$ como:

\begin{equation}
\frac{\textrm{d}f}{\textrm{d}x}=\underset{h\rightarrow0}{\lim}\frac{f\left(x+h,y\right)-f\left(x,y\right)}{h},
\end{equation}

donde se observa que el desplazamiento por $h$ se hace únicamente en el eje $x$, dejando el otro eje intacto. Conceptualmente la derivada parcial respecto a una variable $x$ corresponde al cambio en el funcional en esa dimensión. La evaluación de tal funcional en un punto $\left(x_{0},y_{0}\right)$ viene entonces dada por: 

\begin{equation}
\frac{\textrm{d}f}{\textrm{d}x}_{\left(x_{0},y_{0}\right)}=\underset{h\rightarrow0}{\lim}\frac{f\left(x_{0}+h,y_{0}\right)-f\left(x_{0},y_{0}\right)}{h}
\end{equation}

Por ejemplo, para la función $f\left(x,y\right)=\frac{y^{2}}{b^{2}}-\frac{x^{2}}{a^{2}}$ se tiene que $\frac{\textrm{d}f}{\textrm{d}x}=-\frac{2}{a^{2}}x$ y respecto a $y$ como $\frac{\textrm{d}f}{\textrm{d}y}=\frac{2}{b^{2}}y$.
La evaluación de ambas derivadas parciales en el punto $\left(1,1\right)$
vendrían a ser $\frac{\textrm{d}f}{\textrm{d}x}_{\left(1,1\right)}=-\frac{2}{a^{2}}$
y $\frac{\textrm{d}f}{\textrm{d}y}_{\left(1,1\right)}=\frac{2}{b^{2}}$,
respectivamente. 

En general, para una función con dominio en $\mathbb{R}^{n}$ y codominio en $\mathbb{R}$, $z=f\left(x_{1},\ldots,x_{n}\right)$, la derivada parcial respecto a la variable $x_{i}$ está dada por: 

\begin{equation}
\frac{\textrm{d}f}{\textrm{d}x_{i}}=\underset{h\rightarrow0}{\lim}\frac{f\left(x_{1},\ldots,x_{i}+h,\ldots,x_{n}\right)-f\left(x_{1},\ldots,x_{n}\right)}{h}.
\end{equation}

Veamos ahora la definición formal del $\textbf{vector gradiente}$, primero para una función de con dominio en $\mathbb{R}^{2}$ y codominio en $\mathbb{R}$, $f\left(x,y\right)$ evaluado en cualquier punto $\left(x_{0},y_{0}\right)$:

\begin{equation}
\nabla f_{\left(x_{0},y_{0}\right)}=(\frac{\textrm{d}f}{\textrm{d}x} {\left(x_{0},y_{0}\right)},\frac{\textrm{d}f}{\textrm{d}y}{\left(x_{0},y_{0}\right)})
\end{equation}

y en general para una función con dominio en $\mathbb{R}^{n}$ el vector gradiente en cualquier punto $\left(u_{1},\ldots u_{n}\right)$ viene dado por: 

\begin{equation}
\nabla f_{\left(u_{1},\ldots u_{n}\right)}=(\frac{\textrm{d}f}{\textrm{d}x_{1}} {\left(u_{1},\ldots u_{n}\right)}, \ldots, \frac{\textrm{d}f}{\textrm{d}x_{n}} {\left(u_{1},\ldots u_{n}\right)})
\end{equation}

El vector gradiente denota entonces la dirección para la cual una superficie definida por la función $f$ cambia. Observe que ese vector
cambia de acuerdo al punto $\left(u_{1},\ldots u_{n}\right)$ en el que se evalúe. 

Imagine un punto ubicado en $\left(x_{0},y_{0},z_{0},..\right)$ el vector gradiente $\nabla f_{\left(x_{0},y_{0},z_{0},..\right)}$ en ese punto indica cuál dirección seguir para aumentar el valor de f más rápidamente. El gradiente es perpendicular a las curvas de nivel.

### Ejemplo

Tómese la siguiente función multi-variable $f:\mathbb{R}^{2}\rightarrow\mathbb{R}$, $f\left(x,y\right)=3^{2x}+5^{4y}+2x+y^{4}$. El vector gradiente para una función de dos variables está dado en general por: 

\begin{equation}
\nabla f= (\frac{df}{dx},\frac{df}{dy})
\end{equation}

y en este caso cada derivada parcial está dada por (recordando que para una función $f(x)=a^{x}\Rightarrow f'(x)=a^{x}x'\ln\left(a\right)$):

\begin{equation}
\frac{df}{dx}=2\,3^{2x}\ln\left(3\right)+2
\\
\frac{df}{dy}=4\,5^{4y}\ln\left(5\right)+4y^{3}
\\
\Rightarrow\nabla f= (\left(2\,3^{2x}\ln\left(3\right)+2\right), \left(4\,5^{4y}\ln\left(5\right), 4y^{3}\right))
\end{equation}



# El gradiente matricial

Suponga una función multivariable, la cual toma múltiples entradas (representadas en la matriz $A\in\mathbb{R}^{m\times n}$) y retorna una salida escalar $s\in\mathbb{R}$, por lo que $f:\mathbb{R}^{m\times n}\rightarrow\mathbb{R}$.

El **gradiente** de la función $f$ con respecto a su entrada $A\in\mathbb{R}^{m\times n}$ es la matriz de derivadas parciales
definidas como: 

\begin{equation}
\nabla_{A}f\left(A\right)\in\mathbb{R}^{m\times n}=\begin{bmatrix}\frac{\partial f\left(A\right)}{\partial A_{1,1}} & \frac{\partial f\left(A\right)}{\partial A_{1,2}} & \ldots & \frac{\partial f\left(A\right)}{\partial A_{1,n}}\\
\frac{\partial f\left(A\right)}{\partial A_{2,1}} & \frac{\partial f\left(A\right)}{\partial A_{2,2}} & \ldots & \frac{\partial f\left(A\right)}{\partial A_{2,n}}\\
\vdots & \vdots & \ddots & \vdots\\
\frac{\partial f\left(A\right)}{\partial A_{m,1}} & \frac{\partial f\left(A\right)}{\partial A_{m,2}} & \ldots & \frac{\partial f\left(A\right)}{\partial A_{m,n}}
\end{bmatrix}
\end{equation}

en notación compacta, cada entrada viene dada por: 

\begin{equation}
\left(\nabla_{A}f\left(A\right)\right)_{i,j}=\frac{\partial f\left(A\right)}{\partial A_{i,j}}
\end{equation}

en particular, para una entrada definida en un vector $\vec{x}\in\mathbb{R}^{n}$ el gradiente se define como: 

\begin{equation}
\nabla_{\vec{x}}f\left(\vec{x}\right)=\begin{bmatrix}\frac{\partial f\left(A\right)}{\partial x_{1}}\\
\vdots\\
\frac{\partial f\left(A\right)}{\partial x_{n}}
\end{bmatrix}.
\end{equation}

Es importante remarcar que el **gradiente sólo está definido si la función retorna un escalar**. Esto quiere decir que por ejemplo,
no es posible tomar el gradiente de $A\,\vec{x}$, pues el resultado de tal producto matricial es un vector, y no un escalar.

La derivada matricial parcial es también un operador lineal, tal como la derivada parcial de una función multivariable, por lo que entonces
cumple las propiedades de homogeneidad y superposición:

- $\nabla_{\vec{x}}\left(f\left(\vec{x}\right)+g\left(\vec{x}\right)\right)=\nabla_{\vec{x}}f\left(\vec{x}\right)+\nabla_{\vec{x}}g\left(\vec{x}\right)$
- Para un escalar $s\in\mathbb{R}$,  $\nabla_{\vec{x}}\left(s\,f\left(\vec{x}\right)\right)=s\,\nabla_{\vec{x}}f\left(\vec{x}\right)$


**Se conocen los siguientes gradientes matriciales:**

- $\nabla\left(\vec{x}^{T}\vec{x}\right)=2\,\vec{x}$
- $\nabla\left(\left(A\,\vec{x}\right)^{T}\left(A\,\vec{x}\right)\right)=2\,A\,\vec{x}$
- $\nabla_{\vec{x}}\left(\vec{b}^{T}\vec{x}\right)=\vec{b}$
- $\nabla_{\vec{x}}\left(\vec{x}^{T}\,A\,\vec{x}\right)=2\,A\,\vec{x}$

**Comprobación:**

**(1)** Un ejemplo de una función multidimensional con un vector de entrada es la función $f:\mathbb{R}^{n}\rightarrow\mathbb{R}$

\begin{equation}
        f\left(\vec{z}\right)=\vec{z}^{T}\vec{z}=\sum_{i=1}^{n}z_{i}^{2}    
\end{equation}

la cual como se observa, calcula el producto punto $\vec{z}\cdot\vec{z}$ de su vector entrada 

$\vec{z}=\begin{bmatrix}z_{1}\\
\vdots\\
z_{m}
\end{bmatrix}$ .

Examinando cada una de las $m$ derivadas parciales $\frac{\partial f\left(\vec{z}\right)}{\partial z_{k}}$ (se puede obviar el hecho de que la entrada está dada por un vector y tratar como cualquier función multivariable) se tiene que: 

\begin{equation}
\frac{\partial f\left(\vec{z}\right)}{\partial z_{k}}={\frac{\partial}{\partial z_{k}}z_{1}^{2}}+{\frac{\partial}{\partial z_{k}}z_{2}^{2}}+\ldots+{\frac{\partial}{\partial z_{k}}z_{k}^{2}}+\ldots+{\frac{\partial}{\partial z_{i}}z_{n}^{2}}=2\,z_{k}.
\end{equation}

Donde todos los términos cancelan a cero, excepto $\frac{\partial}{\partial z_{k}}z_{k}^{2}$

Es por ello que el vector gradiente está dado por:

\begin{equation}
\nabla_{\vec{z}}f\left(\vec{z}\right)=\begin{bmatrix}\frac{\partial f\left(\vec{z}\right)}{\partial z_{1}}\\
\vdots\\
\frac{\partial f\left(\vec{z}\right)}{\partial z_{n}}
\end{bmatrix}=\begin{bmatrix}2\,z_{1}\\
\vdots\\
2\,z_{n}
\end{bmatrix}=2\,\vec{z}.
\end{equation}

por lo que entonces el equivalente de la derivada de una función cuadrática de una variable es: 

\begin{equation}
\nabla_{\vec{z}}f\left(\vec{z}\right)=\nabla_{\vec{z}}\left(\vec{z}^{T}\vec{z}\right)=2\,\vec{z}.
\end{equation}


¿Qué sucede si a la entrada de la función $f$ se multiplica por una matriz $A\in\mathbb{R}^{m\times n}$, de modo que se calcule el gradiente $\nabla f\left(A\,\vec{x}\right)$, con $\vec{x}\in\mathbb{R}^{n}$? Se debe interpretar el gradiente de $f$ como la evaluación del mismo en el punto $A\,\vec{x}=\vec{z}$, por lo que entonces el gradiente viene dado por: 

\begin{equation}
\nabla f\left(A\,\vec{x}\right)=\nabla\left(\left(A\,\vec{x}\right)^{T}\left(A\,\vec{x}\right)\right)=2\left(A\,\vec{x}\right)=2\,A\,\vec{x}\in\mathbb{R}^{m}
\end{equation}

Generalizando la función anterior, la cual recibe un vector $\vec{x}\in\mathbb{R}^{n}$ como entrada, y con un vector conocido $\vec{b}\in\mathbb{R}^{n}$:

\begin{equation}
f\left(\vec{x}\right)=\vec{b}^{T}\vec{x}=\sum_{i=1}^{n}b_{i}x_{i}
\end{equation}

con lo que su derivada parcial viene entonces dada por: 

\begin{equation}
\frac{\partial f\left(\vec{x}\right)}{\partial x_{k}}=\frac{\partial}{\partial x_{k}}\sum_{i=1}^{n}b_{i}x_{i}=b_{k}
\end{equation}

Es por ello que se tiene entonce que: 

\begin{equation}
\nabla_{\vec{x}}\left(\vec{b}^{T}\vec{x}\right)=\vec{b}
\end{equation}


Considere ahora la función cuadrática en **(1)** (la cual como ya se vió resulta en un escalar): 

\begin{equation}
f\left(\vec{x}\right)=\vec{x}^{T}\,A\,\vec{x}=\vec{x}^{T}\,\begin{bmatrix}- & \vec{a}_{1,:}^{T} & -\\
- & \vec{a}_{2,:}^{T} & -\\
 & \vdots\\
- & \vec{a}_{m,:}^{T} & -
\end{bmatrix}\,\begin{bmatrix}x_{1}\\
x_{2}\\
\vdots\\
x_{n}
\end{bmatrix}=\begin{bmatrix}x_{1} & \ldots & x_{n}\end{bmatrix}\,\begin{bmatrix}\vec{a}_{1,:}^{T}\:\vec{x}\\
\vec{a}_{2,:}^{T}\:\vec{x}\\
\vdots\\
\vec{a}_{m,:}^{T}\:\vec{x}
\end{bmatrix}=\sum_{i=1}^{n}\sum_{j=1}^{n}A_{i,j}x_{i}x_{j}
\end{equation}

Para calcular la derivada parcial $\frac{\partial f\left(\vec{x}\right)}{\partial x_{k}}$ para cada componente $x_{k}$ del vector de entrada $\vec{x}$, se descomponen las sumatorias anidadas en los casos en que la fila y columna de tal sumatoria es distinta a $k$, en que la fila es igual a $k$, además del caso en que la columna es igual a $k$, y finalmente, cuando se está en la fila y columna $k$:

\begin{equation}
\frac{\partial f\left(\vec{x}\right)}{\partial x_{k}}=\frac{\partial}{\partial x_{k}}\sum_{i=1}^{n}\sum_{j=1}^{n}A_{i,j}x_{i}x_{j}
\\
\Rightarrow\frac{\partial f\left(\vec{x}\right)}{\partial x_{k}}=\frac{\partial}{\partial x_{k}}\left[\sum_{i\neq k}^{n}\sum_{j\neq k}^{n}A_{i,j}x_{i}x_{j}+\sum_{i\neq k}^{n}A_{i,k}x_{i}x_{k}+\sum_{j\neq k}^{n}A_{k,j}x_{k}x_{j}+A_{k,k}x_{k}^{2}\right]
\\\Rightarrow\frac{\partial f\left(\vec{x}\right)}{\partial x_{k}}=\sum_{i\neq k}^{n}A_{i,k}x_{i}+\sum_{j\neq k}^{n}A_{k,j}x_{j}+2A_{k,k}x_{k}=\sum_{i=1}^{n}A_{i,k}x_{i}+\sum_{j=1}^{n}A_{k,j}x_{j}
\end{equation}

Dado que se asume que en la forma cuadrática $A$ es simétrica, lo que implica que $A=A^{T}\Rightarrow A_{i,j}=A_{j,i}$, se tiene que:

\begin{equation}
\Rightarrow\frac{\partial f\left(\vec{x}\right)}{\partial x_{k}}=\sum_{i=1}^{n}A_{i,k}x_{i}+\sum_{j=1}^{n}A_{k,j}x_{j}=2\sum_{i=1}^{n}A_{k,i}x_{i}.
\end{equation}

Es por ello que se concluye que el gradiente de la forma cuadrática está dado por: 
\begin{equation}
\nabla_{\vec{x}}\left(\vec{x}^{T}\,A\,\vec{x}\right)=2\,A\,\vec{x}.
\end{equation}

**Se concluyen entonces los siguientes gradientes matriciales:**

- $\nabla\left(\vec{x}^{T}\vec{x}\right)=2\,\vec{x}$
- $\nabla\left(\left(A\,\vec{x}\right)^{T}\left(A\,\vec{x}\right)\right)=2\,A\,\vec{x}$
- $\nabla_{\vec{x}}\left(\vec{b}^{T}\vec{x}\right)=\vec{b}$
- $\nabla_{\vec{x}}\left(\vec{x}^{T}\,A\,\vec{x}\right)=2\,A\,\vec{x}$



## Mínimos cuadrados
El problema de los mínimos cuadrados en este caso se definirá para encontrar, dada la matriz de rango completo $A\in\mathbb{R}^{m\times n}$, y por ende, invertible y un vector $\vec{b}\in\mathbb{R}^{m\times1},$
el vector $\vec{x}\in\mathbb{R}^{n\times1}$ más cercano al **espacio de columnas** de la matriz $A$, el cual es denotado como $\mathcal{C}\left(A\right)$ y corresponde al espacio generado por las columnas de la matriz $A$, combinadas linealmente por los componentes $x_{i}$ del vector $\vec{x}$: 

\begin{equation}
\mathcal{C}\left(A\right)=\left\{ \vec{v}\in\mathbb{R}^{m}:\vec{v}=A\,\vec{x},\;\vec{x}\in\mathbb{R}^{n},\:A\in\mathbb{R}^{m\times n}\right\} 
\end{equation}

Asumiendo que $A$ es de rango completo y que $n<m$ se tiene que la proyección del vector $\vec{b}\in\mathbb{R}^{m\times1}$ **al cuadrado para simplificar su minimización** en el espacio de columnas
de la matriz $A$ está dado por: 

\begin{equation}
\textrm{proy}\left(\vec{b};A\right)=\textrm{argmin}_{\vec{v}\in\mathcal{C}\left(A\right)}\left\Vert \vec{v}-\vec{b}\right\Vert _{2}^{2}=\textrm{argmin}_{\vec{x}}\left(A\:\vec{x}-\vec{b}\right)\cdotp\left(A\:\vec{x}-\vec{b}\right)
\end{equation}

Es necesario entonces encontrar el vector que minimice la ecuación $\left(A\:\vec{x}-\vec{b}\right)^{T}\left(A\:\vec{x}-\vec{b}\right)$, la cual desarrollándola viene dada por: 

\begin{equation}
\left(A\:\vec{x}-\vec{b}\right)^{T}\left(A\:\vec{x}-\vec{b}\right)=\left(\vec{x}^{T}\,A^{T}-\vec{b}^{T}\right)\left(A\:\vec{x}-\vec{b}\right)=\vec{x}^{T}\,A^{T}A\:\vec{x}-\vec{x}^{T}\,A^{T}\vec{b}-\vec{b}^{T}A\:\vec{x}+\vec{b}^{T}\vec{b}
\end{equation}

Observe con atención los términos $\vec{x}^{T}\,A^{T}\vec{b}$ y $\vec{b}^{T}A\:\vec{x}$. El primer término corresponde a un producto de matrices con dimensiones $1\times n\,n\times m\,m\times1$ lo cual resulta en un escalar, al igual que el segundo término asociado al producto $1\times m\,m\times n\,n\times1$.

Esto quiere decir que al tomar la transpuesta del primer término por ejemplo, el escalar sigue siendo el mismo, por lo que entonces $\left(\vec{x}^{T}\,A^{T}\vec{b}\right)^{T}=\vec{b}^{T}A\:\vec{x}$,

Es por esto que la ecuación anterior se simplifica como sigue: 

\begin{equation}
\left(A\:\vec{x}-\vec{b}=\right)^{T}\left(A\:\vec{x}-\vec{b}\right)=\vec{x}^{T}\,A^{T}A\:\vec{x}-2\,\vec{b}^{T}A\,\vec{x}+\vec{b}^{T}\vec{b}.
\end{equation}

Para realizar la minimización:
\begin{equation}
f\left(\vec{x}\right)=\textrm{argmin}_{\vec{x}}\left(A\:\vec{x}-\vec{b}\right)\cdotp\left(A\:\vec{x}-\vec{b}\right)
\end{equation}

se calculará **el gradiente de tal producto y se igualará a cero**, para encontrar su punto mínimo. Recuerde que para toda función que
resulta en un escalar $f\left(\vec{x}\right)$, calcular el vector gradiente:

\begin{equation}
\nabla_{\vec{x}}\left(A\:\vec{x}-\vec{b}\right)^{T}\left(A\:\vec{x}-\vec{b}\right)=0
\\
\Rightarrow\nabla_{\vec{x}}\left(\vec{x}^{T}\,A^{T}A\:\vec{x}-2\,\vec{b}^{T}A\,\vec{x}+\vec{b}^{T}\vec{b}\right)=0
\\
\Rightarrow\nabla_{\vec{x}}\left(\vec{x}^{T}\,A^{T}A\:\vec{x}\right)-\nabla_{\vec{x}}\left(2\,\vec{b}^{T}A\,\vec{x}\right)+\nabla_{\vec{x}}\left(\vec{b}^{T}\vec{b}\right)=0
\end{equation}

(1) Observe que para el gradiente en el primer término se tiene que $\left(\vec{x}^{T}\,A^{T}A\:\vec{x}\right)=\left(\vec{x}^{T}\,K\:\vec{x}\right)$ con $K=A^{T}A$ una matriz cuadrada, lo que corresponde a la forma cuadrática, para la cual ya se había demostrado que el gradiente viene
dado por: $\nabla_{\vec{x}}\left(\vec{x}^{T}\,K\,\vec{x}\right)=2\,K\,\vec{x}$

(2) Respecto al segundo término se puede reescribir como $\left(2\,\vec{b}^{T}A\,\vec{x}\right)=\left(2\,\vec{k}^{T}\,\vec{x}\right)$, pues observe que del producto $\vec{b}^{T}A$ resulta un vector con valores conocidos $\vec{b}^{T}A=\vec{k}^{T}\in\mathbb{R}^{1\times n}$
dado que los términos del producto se hacen con las dimensiones $1\times m\,m\times n$.

Para una expresión similar, dejando fuera el escalar $2$, ya demostramos que el gradiente viene dado por $\nabla_{\vec{x}}\left(\vec{k}^{T}\vec{x}\right)=\vec{k}$. **Por la regla del gradiente** $\nabla_{\vec{x}}\left(2\,\vec{b}^{T}A\,\vec{x}\right)={(2\vec{b}^{T}A)}^{T}$. 

(3) Finalmente, el tercer término corresponde a una constante, por lo que su gradiente es cero.

**Entonces de (1) y (2) tenemos**: 
\begin{equation}
\Rightarrow2\,A^{T}A\,\vec{x}-{(2\,\vec{b}^{T}A)}^{T}=0
\end{equation}

y tomando la transpuesta del segundo término:

\begin{equation}
\Rightarrow2\,A^{T}A\,\vec{x}-2\,A^{T}\,\vec{b}=0
\\
\Rightarrow A^{T}A\,\vec{x}=A^{T}\,\vec{b}
\\
\Rightarrow\vec{x}=\left(A^{T}A\right)^{-1}A^{T}\,\vec{b}
\end{equation}

De las propiedades de la pseudo-inversa sabemos que $\left(A^{T}A\right)^{-1}A^{T} = A^{+}$

**Entonces** 
\begin{equation}
\Rightarrow\vec{x}=A^{+}\,\vec{b}
\end{equation}
